```python
'''Importao de mdulos'''
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
'''Carregar o conjunto de dados no ambiente''' 
dataset = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv') 
```

```python
'''Visualizao do nmero de registros que reclamam de Atrito'''
dataset.loc[dataset["Attrition"]=="Yes"].count()
```

```python
'''Visualizao do nmero de registros que NO reclamam de Atrito'''
dataset.loc[dataset["Attrition"]=="No"].count()
```

```python
'''Visualizao dos registros que reclamam de Atrito'''
dataset.loc[dataset["Attrition"]=="Yes"]
```

```python
'''Visualizao dos registros que NO reclamam de Atrito'''
dataset.loc[dataset["Attrition"]=="No"]
```

```python
'''
 possvel ver que dentre os que possuem atritos, existe uma predominancia de funcionrios que Viajam 
Raramente. Ser que isso  um fator determinante para o atrito? Para conferir isto, basta gerar um grfico
entre as frequencias de viagem e o nmero de funcionrios.
'''
plt.plot(dataset.loc[dataset["Attrition"]=="Yes"].groupby('BusinessTravel').size().sort_values(), color='red')
plt.plot(dataset.loc[dataset["Attrition"]=="No"].groupby('BusinessTravel').size().sort_values(), color='green')
plt.title('Frequncia de viagem x Nmero de funcionrios')
plt.ylabel('Nmero de Funcionrios')
plt.xlabel('Frequncia com que viajam')
```

```python
''' possvel ver que mesmo havendo uma predominncia de funcionrios que viajam raramente entre os que possuem atrito, o mesmo acontece para os que NO possuem atritos, portanto, este no deve ser um fator determinante.'''
```

```python
'''Visualizao dem grfico da Idade dos Funcionrios'''

plt.plot(dataset.groupby('Age').size())
plt.title('Idade x Quantidade')
plt.ylabel('Quantidade')
plt.xlabel('Idade dos Funcionrios')
```

```python
'''Com a visualizao da idade dos funcionrios x quantidade, agora vamos colocar em um grfico a disperso 
por idade dos funcionrios que reclamaram de atrito para analisar um possvel fator determinante.'''

count = dataset.loc[dataset["Attrition"]=="Yes"].groupby('Age').size().values
idades = dataset.loc[dataset["Attrition"]=="Yes"].groupby('Age').size().index

plt.title('Idade dos funcionrios que reclamaram de Atrito')
plt.pie(count, labels = list(idades))
```

```python
''' possvel ver que o grupo com maior reclamao contm as idades entre 29 e 31 anos.
Agora podemos buscar pela quantidade exata de reclamantes nessas idades'''

dataset.loc[(dataset["Attrition"]=="Yes") & (dataset["Age"] >= 29) & (dataset["Age"] <=31)].groupby('Age').size().values
```

```python
dataset.loc[(dataset["Attrition"]=="No") & (dataset["Age"] >= 29) & (dataset["Age"] <=31)].groupby('Age').size().values
```

```python
''' possvel ver que mesmo nas idades que houveram mais reclamaes, ainda sim existe uma quantidade bem maior
de funcionrios nessas mesmas idades que no reclamaram. Portanto, este tambm no pode ser um fator determinante'''

'''Agora iremos analisar a relao de Atritos e Departamentos'''
dataset.groupby('Department').size().sort_values()
```

```python
dataset.loc[dataset["Attrition"]=="Yes"].groupby('Department').size().sort_values()
```

```python
dataset.loc[dataset["Attrition"]=="No"].groupby('Department').size().sort_values()
```

```python
'''Anlise da disperso dos funcionrios que reclamaram pela distncia de suas casas ao trabalho'''

count = dataset.loc[dataset["Attrition"]=="Yes"].groupby('DistanceFromHome').size().values
distancia = dataset.loc[dataset["Attrition"]=="Yes"].groupby('DistanceFromHome').size().index

plt.title('Distancia de Casa dos funcionrios que reclamaram de Atrito')
plt.pie(count, labels = list(distancia))
```

```python
'''Anlise da disperso dos funcionrios que NO reclamaram pela distncia de suas casas ao trabalho'''

count = dataset.loc[dataset["Attrition"]=="No"].groupby('DistanceFromHome').size().values
distancia = dataset.loc[dataset["Attrition"]=="No"].groupby('DistanceFromHome').size().index

plt.title('Distancia de Casa dos funcionrios que NO reclamaram de Atrito')
plt.pie(count, labels = list(distancia))
```

```python
''' possvel ver que mesmo havendo uma predominncia de funcionrios que reclamaram morando h 2km do trabalho,
existem funcionrios que moram na mesma distncia e no reclamaram'''

'''Anlise dos Campos de Estudo x Departamentos de atuao'''

dataset.groupby('EducationField').size().sort_values()
```
