```python
'''
É possível ver que dentre os que possuem atritos, existe uma predominancia de funcionários que Viajam 
Raramente. Será que isso é um fator determinante para o atrito? Para conferir isto, basta gerar um gráfico
entre as frequencias de viagem e o número de funcionários.
'''
plt.plot(dataset.loc[dataset["Attrition"]=="Yes"].groupby('BusinessTravel').size().sort_values(), color='red')
plt.plot(dataset.loc[dataset["Attrition"]=="No"].groupby('BusinessTravel').size().sort_values(), color='green')
plt.title('Frequência de viagem x Número de funcionários')
plt.ylabel('Número de Funcionários')
plt.xlabel('Frequência com que viajam')
```

```python
'''É possível ver que mesmo havendo uma predominância de funcionários que viajam raramente entre os que possuem atrito, o mesmo acontece para os que NÃO possuem atritos, portanto, este não deve ser um fator determinante.'''
```

```python
'''Visualização dem gráfico da Idade dos Funcionários'''

plt.plot(dataset.groupby('Age').size())
plt.title('Idade x Quantidade')
plt.ylabel('Quantidade')
plt.xlabel('Idade dos Funcionários')
```

```python
'''Com a visualização da idade dos funcionários x quantidade, agora vamos colocar em um gráfico a dispersão 
por idade dos funcionários que reclamaram de atrito para analisar um possível fator determinante.'''

count = dataset.loc[dataset["Attrition"]=="Yes"].groupby('Age').size().values
idades = dataset.loc[dataset["Attrition"]=="Yes"].groupby('Age').size().index

plt.title('Idade dos funcionários que reclamaram de Atrito')
plt.pie(count, labels = list(idades))
```

```python
'''É possível ver que o grupo com maior reclamação contém as idades entre 29 e 31 anos.
Agora podemos buscar pela quantidade exata de reclamantes nessas idades'''

dataset.loc[(dataset["Attrition"]=="Yes") & (dataset["Age"] >= 29) & (dataset["Age"] <=31)].groupby('Age').size().values
```

```python
dataset.loc[(dataset["Attrition"]=="No") & (dataset["Age"] >= 29) & (dataset["Age"] <=31)].groupby('Age').size().values
```

```python
'''É possível ver que mesmo nas idades que houveram mais reclamações, ainda sim existe uma quantidade bem maior
de funcionários nessas mesmas idades que não reclamaram. Portanto, este também não pode ser um fator determinante'''

'''Agora iremos analisar a relação de Atritos e Departamentos'''
dataset.groupby('Department').size().sort_values()
```

```python
dataset.loc[dataset["Attrition"]=="Yes"].groupby('Department').size().sort_values()
```

```python
dataset.loc[dataset["Attrition"]=="No"].groupby('Department').size().sort_values()
```

```python
'''Análise da dispersão dos funcionários que reclamaram pela distância de suas casas ao trabalho'''

count = dataset.loc[dataset["Attrition"]=="Yes"].groupby('DistanceFromHome').size().values
distancia = dataset.loc[dataset["Attrition"]=="Yes"].groupby('DistanceFromHome').size().index

plt.title('Distancia de Casa dos funcionários que reclamaram de Atrito')
plt.pie(count, labels = list(distancia))
```

```python
'''Análise da dispersão dos funcionários que NÃO reclamaram pela distância de suas casas ao trabalho'''

count = dataset.loc[dataset["Attrition"]=="No"].groupby('DistanceFromHome').size().values
distancia = dataset.loc[dataset["Attrition"]=="No"].groupby('DistanceFromHome').size().index

plt.title('Distancia de Casa dos funcionários que NÃO reclamaram de Atrito')
plt.pie(count, labels = list(distancia))
```

```python
'''É possível ver que mesmo havendo uma predominância de funcionários que reclamaram morando há 2km do trabalho,
existem funcionários que moram na mesma distância e não reclamaram'''

'''Análise dos Campos de Estudo x Departamentos de atuação'''

dataset.groupby('EducationField').size().sort_values()
```
